```python
import math
x = np.random.randint(30,size=20)
y = np.random.randint(30, size=20)
z = 18
m = 20
plt.plot(z,m,'^r')
plt.plot(x[:10],y[:10],'bo')
plt.plot(x[10:],y[10:],'yo')

plt.show()

d = []
for i in range(len(x)):
    m = (x[i] - z)** 2 + (y[i] -m )** 2
    m = math.sqrt(m)
    d.append(m)
    
# for i in range(len(d)):
#     print(i,d[i])

index = np.argsort(d)
print(index)


k = 9
xanh = 0
vang = 0
for i in range(k):
    if index[i] < 10:
        xanh += 1
    else:
        vang += 1
        
print('xanh = ',xanh)
print('vang = ',vang)
```
