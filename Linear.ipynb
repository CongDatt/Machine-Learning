```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7]
y = [2, 4.2, 6.5 ,10.3 ,10.5 , 12.2 ,14.1]
w = [4,1]
plt.plot(x,y, 'ro')
plt.xlabel('X')
plt.ylabel('Y')
x0 = np.linspace(start=1, stop=10, num=50)
y0 = w[0]+w[1]*x0

plt.plot(x0,y0)
plt.show()
```

```python
# w : tham so
def f(x_i,w):
    return w[0]+w[1]*x_i

def loss(x,y,w):
    error = 0
    for i in range(len(x)):
        error += (y[i] - f(x[i], w))**2
    return error/(2*len(x))

# tinh dao ham tao diem w[0], w[1]
def derivative(x,y,w):
    d0 =0 #w[0]
    d1 =0 #w[1]
    for i in range(len(x)):
        d1 += x[i]*(f(x[i],w) - y[i])
        d0 = f(x[i],w) - y[i]
    return d0/len(x),d1/len(x)
```

```python
epoch = 10
learning_rate = 0.01
w = [1,1]
loss_old = loss(x,y,w)

for i in range(epoch):
    plt.plot(x,y,'ro')
    plt.xlabel('X')
    plt.ylabel('Y')   
    x0 = np.linspace(start=1, stop=10, num=50)
    y0 = w[0]+w[1]*x0
    plt.plot(x0,y0)
    plt.show()
    
    a,b = derivative(x,y,w)
    w[0]  = w[0] - a*learning_rate
    w[1]  = w[1] - b*learning_rate    
    
    loss_new = loss(x,y,w)
    print('epoch_',i,':')
    print('w=',w)
    if abs(loss_new - loss_old) < 0.0001:
        break
    loss_old = loss_new
```
