```python
#3. cho một mảng gồm n chữ cái, được sinh ngẫu nhiên
# hãy đếm frequency của các chữ cái
# vẽ đồ thị bar cho các chữ cái và tần suất tương ứng

import string
string.ascii_letters
freq = []
for i in range(20):
    a = random.choice(string.ascii_lowercase)
    freq.append(a)

elements_count = {}
for element in freq:
   if element in elements_count: # check exsit or not ?
      elements_count[element] += 1 # if yes, +1
   else:
      elements_count[element] = 1
        
labels = []
sizes = []
for key, value in elements_count.items():
    labels.append(key)
    sizes.append(value)
    print(f"{key}: {value}")    



plt.pie(sizes, labels=labels, explode=None, colors=None,shadow=True,autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('frequency of Random Alphabets')
plt.show()
```

```python
#5. Vẽ đồ thị của hàm số z = 2x + 3y
# trên không gian 3 chiều 

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
ax = plt.axes(projection="3d")

ax.set_title('Test')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')



x = [1, 2, 3, 4, 5]
y = [1,2, 3, 4 ,5]
z = []

for i in range(len(x)):
    a = 2*x[i] + 3*y[i]
    z.append(a)

    print(z)
    
    

ax.scatter(x, y, z, color='r')

plt.xlim(0,15)
plt.ylim(0,15)
ax.set_zlim(0,25)

plt.show()
```
