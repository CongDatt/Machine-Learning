```python
import matplotlib.pyplot as plt
import numpy as np 
```

Việt:
- matplotlib là thư viện hoàn hảo dùng cho vẽ đồ thị (cho visualization nói chung) ứng dụng đặc biệt cho phân tích dữ liệu và cho học máy. 
- Module pyplot bao gồm những chức năng hữu dụng nhất, thường viết tắt bằng plt 

English:
- Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.
- Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias.


```python
#1. thiết lập tham số các trục của đồ thị
# đặt tên cho trục và tên của đồ thị 
plt.title("Test1")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.show()
```

```python
# vẽ với scale trên trục 
plt.xlim(0,20)
#plt.xlim(5,11)
plt.ylim(0,200)
plt.xlabel("X label")
plt.show()
```

```python
#2. vẽ các điểm trên đồ thị 
# (x1,y1), (x2,y2), ...,(xn,yn)
# x lưu trong mảng X
# y lưu trong mảng Y

x = [4,2,7,5]
y = [2,15,8,5]
#x = np.array(x)
plt.scatter(x,y,marker='^', color='b')
plt.show()
```

```python
# vẽ đường trên đồ thị 
x = [2,5]
y = [4,8]
plt.xlim(0,10)
plt.ylim(0,10)
#plt.plot(x,y)
plt.plot(x,y,marker='o',color='b',ls='-')
plt.show() 
```

```python
# vẽ đồ thị y = x^2
x=[i for i in range(-10,11)]
y=[i**2 for i in x ]
print(x)
print(y)
plt.plot(x,y)
plt.show()
```

```python
# vẽ đồ thị y = 2x+anpha
x=[i for i in range(1,10)]
y=[2*i+4*np.random.rand() for i in x ]
print(x,y)
plt.plot(x,y,marker='o',color='r')
plt.show()
```

```python
# vẽ hai đồ thị trên một trục toạ độ 
x = [-5,8]
y = [2,10]
plt.xlim(-10,15)
plt.ylim(-2,15)
plt.plot(x,y)
x1=[i for i in range(-3,4)]
y1=[i**2 for i in x1 ]
plt.plot(x1,y1,color='r')

plt.show() 
```

```python
#3. vẽ đồ thị bar thể hiện sự phân bố dữ liệu 
names = ['weak','normal','good','excellent']
values = [20,30,35,18]

plt.figure(figsize=(5, 4))
#plt.subplot(1,2,1)
plt.bar(names, values)
plt.show()
```

```python

labels = 'A', 'B', 'C', 'D'
sizes = [40, 20, 20, 20]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (20, 20, 20, 20)
#plt.pie(sizes, explode=explode, labels=labels, colors=colors,
#        autopct='%1.1f%%', shadow=True, startangle=90)
plt.pie(sizes, labels=labels,explode=explode, colors=colors,shadow=True,autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Year 2013')
plt.show()
```

```python
# working with categorial variable 
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(15, 3))

plt.subplot(1,3,1)
plt.bar(names, values)

plt.subplot(1,3,2)
plt.scatter(names, values)

plt.subplot(1,3,3)
plt.plot(names, values)

plt.suptitle('Categorical Plotting')
plt.show()
```

```python
#4. Đồ thị 3D 
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
#ax = Axes3D(fig)
ax = plt.axes(projection="3d")

ax.set_title('Test')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

x = [1,3,4,5,7,9,5]
y = [2,3,5,5,6,7,5]
z = [6,4,3,7,8,20,10]

ax.scatter(x, y, z,color='r')
plt.xlim(0,15)
plt.ylim(0,15)
ax.set_zlim(0,25)

plt.show()
```
